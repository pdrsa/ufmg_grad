```python
def plotting(result):
    algorithm = result["algorithm"]
    learning_rate = result["learning_rate"]
    hidden_size = result["hidden_size"]
    train_losses = result["results"][0]
    test_acc = result["results"][1]
    
    subtitle = "Loss and Accuracy"
    title = f"{subtitle} with LR = {learning_rate} and Hidden Size = {hidden_size}"
    plt.title(title)
    
    plt.plot([str(i+1) for i in range(len(train_losses))], train_losses, color=c[algorithm][0])
    plt.plot([str(i+1) for i in range(len(test_acc))], [x/100 for x in test_acc], color=c[algorithm][1])
    plt.legend(['Train Loss', 'Test Accuracy'], loc='upper right')
    plt.xlabel('Epoch')
    plt.ylabel("Values")
    
    os.makedirs(subtitle, exist_ok=True)
    subfolder = os.path.join(subtitle, algorithm)
    os.makedirs(subfolder, exist_ok=True)
    path = os.path.join(subfolder, f'lr_{learning_rate}_hs_{hidden_size}.png')
    plt.savefig(path, dpi=1200)
    plt.close()
```

```python
from tqdm import tqdm
c = {
    "stochastic": ["blue", "black"],
    "gradient_descent": ["red", "green"],
    "mini_batch": ["yellow", "purple"]
}

for result in tqdm(results):
    plotting(result)
```
