### Tarefa 1 - a)

Primeiro, iremos criar uma função que, dado um número N de passos, retornas as tuplas de (Tempo, Posição) de uma caminhada aleatória de N passos, com cada passos distribuído aleatóriamente num intervalo que também é recebido como parâmetro.

```python
import numpy as np

def rand(a, b):
    return (b - a) * np.random.random_sample() + a

# Interval should be (a, b) with a being the start and b the end of the interval
def randomWalk1D(nSteps, interval):
    a, b = interval
    steps = [rand(a, b) for i in range(nSteps)]
    tuples = [[0, 0]]
    for i in range(nSteps):
        tuples.append([i, tuples[i][1] + steps[i]])
        
    return np.array(tuples)

def randomWalk2D(nSteps, interval):
    a, b = interval
    steps = [[rand(a, b), rand(a, b)] for i in range(nSteps)]
    tuples = [[0, 0, 0]]
    for i in range(nSteps):
        tuples.append([i, tuples[i][1] + steps[i][0], tuples[i][2] + steps[i][1]])
        
    return np.array(tuples)
```

Agora iremos plotar alguns gráficos Tempo x Posição para algumas caminhadas com $N = 10000$ e intervalo $[-\frac{1}{2}, \frac{1}{2}]$.

```python
# Customizing some colors
colors = ["#281641", "#265E57", "#AB3433", "#F5B14C", "#008000"]

import matplotlib.pyplot as plt
for i in range(5):
    tuples = randomWalk1D(10000, (-1/2, 1/2))
    plt.plot(tuples[:,0], tuples[:,1], color = colors[i])
    plt.title("Caminhada Aleatória de 10000 passos em (-1/2, 1/2)")
    plt.xlabel("Tempo")
    plt.ylabel("Posição")
    plt.grid()
    plt.show()
```

### Tarefa 2

Primeiramente, iremos definir uma função que recebe $W$ e $N$ como parâmetros e nos retorna um vetor com as posições finais de $W$ caminhadas aleatórias de $N$ passos. O intervalo para os passos também será recebido como parâmetro.

```python
def finalPositions(nWalks, nSteps, interval):
    pos = []
    for i in range(nWalks):
        walk = randomWalk2D(nSteps, interval)
        pos.append([walk[-1][1], walk[-1][2]])
    
    return np.array(pos)
```

Agora iremos plotar os pontos finais de $10000$ caminhadas com $N = 1$ e $N = 10$

```python
N = 1
positions = finalPositions(10000, N, (-2, 2))

plt.scatter(positions[:,0], positions[:,1], color = colors[2])
plt.xlim(-20, 20)
plt.ylim(-20, 20)
plt.title("Caminhadas com N = 1")
ax = plt.gca() 
ax.set_aspect(1.0)
plt.show()

N = 10
positions = finalPositions(10000, N, (-2, 2))

plt.scatter(positions[:,0], positions[:,1], color = colors[2])
plt.xlim(-20, 20)
plt.ylim(-20, 20)
plt.title("Caminhadas com N = 10")
ax = plt.gca() 
ax.set_aspect(1.0)
plt.show()
```
