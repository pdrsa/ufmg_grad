### Tarefa 3 - Integração Monte Carlo

# Requisitos

```python
import numpy as np
import matplotlib.pyplot as plt
from random import uniform
from math import e, inf, sin, pi
```

# Funções

```python
def generate_points(n, width, height):
    ret = []
    for _ in range(n):
        x = uniform(0, width)
        y = uniform(0, height)
        ret.append((x,y))
    return ret

def find_max(a, b, f):
    max_ = -inf
    interval = np.linspace(a,b,100)
    for number in interval:
        res = f(number)
        if res > max_:
            max_ = res
    return max_

def find_min(a, b, f):
    min_ = inf
    interval = np.linspace(a,b,100)
    for number in interval:
        res = f(number)
        if res < min_:
            min_ = res
    return min_

def monte_carlo(n,a,b,f):
    """
    n é o número de pontos gerados
    a e b são os limites do intervalo
    f é a função
    """
    width = b-a
    height = find_max(a,b,f)
    area = width*height

    points = generate_points(n,width,height)
    all_points = len(points)
    
    points_accepted = []
    points_not      = []
    for point in points:
        # if y < f(x):
        if point[1] < f(point[0]):
            points_accepted.append(point)
        else:
            points_not.append(point)
            
    return [area * float(len(points_accepted)) / all_points, np.array(points_accepted), np.array(points_not)]
```

# Questões


### 1.


```python
def f(x):
    return 1-pow(x,2)

n = 100000
a = 0
b = 1
print(f'O que achamos: {monte_carlo(n,a,b,f)[0]}')
print(f'Resultado analítico: {2/3}')
```

```python
N = [100, 1000, 10000]
p1 = [f(x) for x in np.linspace(a, b, 10000)]
for n in N:
    val, p2, p3 = monte_carlo(n, a, b, f)
    title = "N = " + str(n) + " _ Valor encontrado para a integral: " + str(val)
    plt.title(title)
    plt.plot(np.linspace(a, b, 10000), p1, 'g', color = 'black')
    plt.scatter(p2[:,0], p2[:, 1], color = 'blue', s=8)
    plt.scatter(p3[:,0], p3[:, 1], color = 'orange', s=8)
    plt.show()
```

```python
repeticoes = 10000
N = [100, 1000, 10000]

min_ = find_min(a,b,f)
max_ = find_max(a,b,f)

for n in N:
    l = []
    for x in range(repeticoes):
        l.append(monte_carlo(n,a,b,f)[0])
    plt.hist(l)
    plt.xlim(min_, max_)
    plt.title(f'N total = {n}')
    plt.show()
```

# Comentário:

Foi possível observar que utilizando o método apresentado, encontrou-se valores bem próximos
dos valores reais de cada uma das integrais. Também foi observado, utilizando os gráficos gerados
para os diferentes valores de N, que quanto maior for o N, mais o valor encontrado ficou próximo
do valor real. Isso é evidenciado pela largura das distribuições dos valores nos histogramas, que
é bem menor quando o valor de N é grande.
Concluímos então, baseado nos resultados obtidos, que com um N suficientemente grande foi
possível encontrar um valor aproximado para as integrais apresentadas utilizando o método de
Monte Carlo. Também observou-se que dependendo do valor de N escolhido, o nível de precisão
varia, sendo que N maiores tendem a garantir uma maior precisão. Apesar de aumentar esse
valor de N ser uma boa maneira de diminuir o erro encontrado na resposta final, existem outras
maneiras de identificar e tratar esse potencial erro.
Uma maneira de diminuir esse erro pode ser repetir o processo de gerar diferentes versões do
método de Monte Carlo para um determinado problema, utilizando o mesmo valor de N, e fazer
a média dos valores encontrados nessas repetições, por exemplo. Dessa forma, espera-se que o
erro gerado pela aleatoriedade dos pontos diminua e a precisão do valor encontrado aumente em
relação ao valor real.
